```python
import matplotlib.pyplot as plt
import numpy as np

'''
The Heaviside step function.

'''

def Heaviside(t):
  unit_step = numpy.arange(t.shape[0])
  lcv = numpy.arange(t.shape[0])
  for place in lcv:
    if t[place] == 0:
      unit_step[place] = .5
    elif t[place] > 0:
      unit_step[place] = 1
    elif t[place] < 0:
      unit_step[place] = 0
  return unit_step  
```

```python
t = np.linspace(-5,5,100)
fig,ax = plt.subplots(figsize=(6,3), dpi=500) 
ax.plot(t,Heaviside(t), c='green',marker="o" )
ax.grid()
ax.set_ylim(-0.1,1.1)
```
